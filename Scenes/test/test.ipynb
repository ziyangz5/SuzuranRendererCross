```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
DMG = [170,350,340]
sDMG = 500
MAX_SKILL_NUM = 10000
tDelta = 20
```

```python
totalDMGList1 = [] #历史记录：截至目前这个技能，一共造成了多少伤害
totalDMG1 = 0
tamashi = 0
falled_tamashi = 0
r = 0
i = 0
while (r < MAX_SKILL_NUM):
    totalDMG1 += DMG[i] #1 2 3 循环
    totalDMGList1.append(totalDMG1)#记录截止到目前这个技能造成的总伤害
    if (i == 2):
        tamashi += tDelta #魂+10
    i += 1
    r += 1
    i %= 3
    if (tamashi == 100):
        tamashi = 0
        r += 1
        totalDMG1 += sDMG
        totalDMGList1.append(totalDMG1) 
        r += 1
        totalDMG1 += sDMG
        totalDMGList1.append(totalDMG1) 
```

```python
totalDMGList2 = []
totalDMG2 = 0
tamashi = 0
falled_tamashi = 0
falled_tamashi_list = []
r = 0
i = 0
while (r < MAX_SKILL_NUM):
    totalDMG2 += DMG[i]#1 2 3 循环
    totalDMGList2.append(totalDMG2)#记录截止到目前这个技能造成的总伤害
    falled_tamashi_list.append(falled_tamashi)#记录截止到目前这个技能一共亏了多少魂
    if (i == 2):
        tamashi += tDelta#魂+10
    i += 1
    r += 1
    i %= 3
    if (tamashi == 120):
        tamashi = 0
        
        falled_tamashi += 20
        falled_tamashi_list.append(falled_tamashi)
        r += 1
        totalDMG2 += sDMG
        totalDMGList2.append(totalDMG2) 
        falled_tamashi_list.append(falled_tamashi)
        r += 1
        totalDMG2 += sDMG
        totalDMGList2.append(totalDMG2) 
```

```python
plt.plot(totalDMGList1)
plt.plot(totalDMGList2)
```

```python
l1 = np.array(totalDMGList1)
l2 = np.array(totalDMGList2)
result = (l1-l2)/np.array(falled_tamashi_list)*10
plt.plot(result) #（循环1-循环2）/（亏掉的魂数）*10
```
